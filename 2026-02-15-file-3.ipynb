```python
playing = True

while playing:
    choice = input('Do you want to use the calculator? (y/n): ').lower()

    if choice == 'y':
        print('1 - add\n2 - subtract\n3 - multiply\n4 - divide')

        try:
            option = int(input('Choose an operation: '))

            if option in [1, 2, 3, 4]:
                num1 = float(input('Enter the first number: '))
                num2 = float(input('Enter the second number: '))

                if option == 1:
                    result = num1 + num2
                elif option == 2:
                    result = num1 - num2
                elif option == 3:
                    result = num1 * num2
                elif option == 4:
                    if num2 == 0:
                        print("You can't divide by zero!")
                        continue
                    result = num1 / num2

                print('Result:', result)

            else:
                print('Invalid operation.')

        except ValueError:
            print('Please enter numbers only.')

    elif choice == 'n':
        print('Goodbye ')
        playing = False

    else:
        print("Please type 'y' or 'n'.")
```
