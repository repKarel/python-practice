```python
import random

number_to_guess = random.randint(1, 100)

low_count = 0
high_count = 0

while True:
    try:
        guess = int(input("Guess the number between 1 and 100: "))

        if guess < number_to_guess:
            low_count += 1
            high_count = 0

            if low_count >= 2:
                print("Still too low Try going higher")
            else:
                print("Too low")

        elif guess > number_to_guess:
            high_count += 1
            low_count = 0

            if high_count >= 2:
                print("Still too high! Try going lower")
            else:
                print("Too high")

        else:
            print("CongratulationsYou got it right!")
            break

    except ValueError:
        print("Please enter a valid number")
```
